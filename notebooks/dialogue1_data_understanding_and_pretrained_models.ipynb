```python
def hfds2df(hfds):
  '''hugging face dataset to dataframe'''
  if 'train' in hfds:
    dfs = []
    for split, hfds1 in hfds.items():
      df = hfds1.to_pandas()
      df['split'] = split
      dfs.append(df)    
    return pd.concat(dfs)  
  else:
    return hfds.to_pandas()
```

```python
## dialogpt

from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_name = 'microsoft/DialoGPT-large'
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(model_name)
```

```python
def interactive_talker(history_collection=None):
  history_collection = history_collection if history_collection is not None else []
  print('starting conversation')
  display(history_collection)
  def model_speak(user_input):
    history_collection.append(user_input)
    encodable_input = str(tokenizer.eos_token).join(history_collection) + tokenizer.eos_token
    input_ids = tokenizer.encode(encodable_input, return_tensors='pt')
    generated_by_model_ids = model.generate(input_ids, max_length=1000, pad_token_id=tokenizer.eos_token_id)
    n = input_ids.shape[-1]
    output_text = tokenizer.decode(generated_by_model_ids[0, n:], skip_special_tokens=True)
    history_collection.append(output_text)
    return output_text

  return model_speak


# talker = interactive_talker()

# a = talker('Hi man! how are you doing?')
```

```python
# list of preconverstaion inferencing

def inference_on_conversation(conversation_list, n=1):
  initials = conversation_list[:2*n]
  conversation_list = conversation_list[2*n:]

  talker = interactive_talker(initials)

  for conv, expected_response in zip(conversation_list[::2], conversation_list[1::2]):
    print(f'user: {conv}')
    print(f'dgpt: {talker(conv)}')
    print(f'true: {expected_response}')
    print()
```

```python
def is_onoff(l):
  if not l: return True
  if len(l) == 1: return True
  return bool(int(l[0])) != bool(int(l[1])) and is_onoff(l[1:])

onoff_set = cgl_df[cgl_df.agent_turn.apply(lambda x: is_onoff(x.tolist()))]
shortlisted = onoff_set[onoff_set.agent_turn.apply(lambda x: len(x)>11)]

convs = shortlisted.utterance.sample(20, random_state=414).values.tolist()
convs[0]
```
